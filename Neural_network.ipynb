```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('cancer.csv')
print(data)
```

```python
X = data.iloc[:,:8]
y = data.iloc[:,9]
#When you read data into  y you will observe that y are rank 1 arrays. rank 1 array will have a shape of (m, ) where as rank 2 arrays will have a shape of (m,1). When operating on arrays its good to convert rank 1 arrays to rank 2 arrays because rank 1 arrays often give unexpected results.
#To convert rank 1 to rank 2 array we use someArray[:,np.newaxis].
y = y[:,np.newaxis]
print(X)
print(y)
```

```python
#Mean Normalization
X = (X-np.mean(X))/np.std(X)
print(X)
```

```python
#Activation function
def sigmoid(s):
    return 1/(1+np.exp(-s))
```

```python
def sigmoid_derv(s):
    return s*(1-s)
```

```python
epochs = 5000 #Number of iterations
alpha = 0.1 #learning rate
input_nuerons = X.shape[1] #number of featuers in dataset
hidden_layers = 16 #number of hidden layer neurons
output = 1 #number of layers in output
```

```python
#Variable initialization
weights = np.random.uniform(size=(input_nuerons,hidden_layers))
bias = np.random.uniform(size=(1,hidden_layers))
weights_out = np.random.uniform(size=(hidden_layers,output))
bias_out = np.random.uniform(size=(1,output))
```

```python
for i in range(epochs):
    #Forward propagation
    hidden_input1 = np.dot(X,weights)
    hidden_input1 = hidden_input1+bias
    activations1 = sigmoid(hidden_input1)
    hidden_input2 = np.dot(activations1,weights_out)
    hidden_input2 = hidden_input2 + bias_out
    out = sigmoid(hidden_input2)
    
    
    #Backward propagation
    error = y - out
    gradient_output = sigmoid_derv(out)
    gradient_hidden = sigmoid_derv(activations1)
    output_2 = error*gradient_output
    error_hidden = np.dot(output_2,weights_out.T)
    output_3 = error_hidden * gradient_hidden
    weights_out += np.dot(activations1.T,output_2) *alpha
    bias_out += np.sum(output_2, axis=0) *alpha
    weights += np.dot(X.T,output_3) *alpha
    bias += np.sum(output_3, axis=0) *alpha
print(weights)
print (out)
```
